    'C': [1, np.nan, np.nan, 4, 5]
})

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a sample DataFrame with missing values


# Plot the number of missing values before filling
plt.figure(figsize=(10, 5))
df.isnull().sum().plot(kind='bar', title='Missing Values Before Filling')
plt.show()

# Fill missing values with the mean of each column
df_filled = df.apply(lambda col: col.fillna(col.mean()), axis=0)

# Plot the number of missing values after filling
plt.figure(figsize=(10, 5))
df_filled.isnull().sum().plot(kind='bar', title='Missing Values After Filling')
plt.show()
```
